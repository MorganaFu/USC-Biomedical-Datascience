With your local Git repository now complete, follow the next steps to push it to GitHub. Start by logging into GitHub and creating an empty GitHub repo with your GitHub account. It can be named whatever you want it to be.

Next, create a set of public & private SSH keys to access your GitHub repo from this Colab notebook. This will involve following the steps outlined in the `Following GitHub secruity protocol with SSH keys` section. You may use multiple code-blocks to do this.

Generate a new private key ("key").

```python
%%bash
ssh-keygen -t ed25519 -C '[jiaqifu@usc.edu]' -f ~/.ssh/id_ed25519 -P ''
cat > ~/.ssh/config
Host *
  AddKeysToAgent yes
  IdentityFile ~/.ssh/id_ed25519
```

Install the private key ("key").

```python
%%bash
eval "$(ssh-agent -s)"
ssh-add ~/.ssh/id_ed25519
```

Generate and register a public key ("keyhole").

```python
%%bash
cat ~/.ssh/id_ed25519.pub
```

```python
%%bash
ssh -o "StrictHostKeyChecking no" -T git@github.com | exit 0
```
